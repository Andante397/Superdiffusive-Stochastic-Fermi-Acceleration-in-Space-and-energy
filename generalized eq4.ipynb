```python
import numpy as np
import matplotlib.pyplot as plt
from numba import njit

# ============================================================
# 0. Switches and parameters
# ============================================================

USE_REJECTION = True

SEED = 42

L_BOX = 1.0e10
N_GRID = 600
SCATTER_DENSITY = 0.10

N_PARTICLES = 50000

N_STEPS = 50000
RECORD_EVERY = 50

T_INIT_EV = 100.0

TMIN_R = 0.5
TMIN_E = 1.0
TMAX_FIT = 7.0

N_BINS_3C = 16
MIN_BIN_3C = 20

MAX_EVENTS_PER_PARTICLE = 300000

MAX_EVENT_SAMPLES = 1_000_000

# ============================================================
# 1. Constants
# ============================================================

C = 3.0e10
M_E = 9.11e-28
M_E_C2 = 8.187e-7
V_A = 7.0e8

ERG_TO_EV = 6.242e11
EV_TO_ERG = 1.0 / ERG_TO_EV


# ============================================================
# 1.5 Digitized article data
# ============================================================
# 在这里保留你原来的三段数据：
#
# ARTICLE_SPATIAL_MSD = np.array([...], dtype=float)
# ARTICLE_ENERGY_MSD = np.array([...], dtype=float)
# ARTICLE_MEAN_DW = np.array([...], dtype=float)
#
# ============================================================


# ============================================================
# 2. Non-numba utilities
# ============================================================

def fit_power(t, y, tmin, tmax):
    mask = (
        (t >= tmin)
        & (t <= tmax)
        & (t > 0)
        & (y > 0)
        & np.isfinite(t)
        & np.isfinite(y)
    )

    if np.sum(mask) < 5:
        return np.nan, lambda x: np.full_like(x, np.nan, dtype=float)

    a, b = np.polyfit(np.log(t[mask]), np.log(y[mask]), 1)
    return a, lambda x: np.exp(b) * x**a


def fit_article_data(article_data):
    t = article_data[:, 0]
    y = article_data[:, 1]
    return fit_power(t, y, t.min(), t.max())


def make_scatter_grid_flat(N_grid, scatter_density, rng, chunk_size=5_000_000):
    total = N_grid**3
    flat = np.empty(total, dtype=np.bool_)

    for start in range(0, total, chunk_size):
        end = min(start + chunk_size, total)
        flat[start:end] = rng.random(end - start) < scatter_density

    return flat


# ============================================================
# 3. Numba helper functions
# ============================================================

@njit
def dir_components(dir_id):
    if dir_id == 0:
        return 1.0, 0.0, 0.0
    if dir_id == 1:
        return -1.0, 0.0, 0.0
    if dir_id == 2:
        return 0.0, 1.0, 0.0
    if dir_id == 3:
        return 0.0, -1.0, 0.0
    if dir_id == 4:
        return 0.0, 0.0, 1.0
    return 0.0, 0.0, -1.0


@njit
def dir_int_components(dir_id):
    if dir_id == 0:
        return 1, 0, 0
    if dir_id == 1:
        return -1, 0, 0
    if dir_id == 2:
        return 0, 1, 0
    if dir_id == 3:
        return 0, -1, 0
    if dir_id == 4:
        return 0, 0, 1
    return 0, 0, -1


@njit
def nearest_lattice_dir_id(vx, vy, vz):
    """
    Project a lab-frame outgoing velocity vector onto the nearest
    one of six lattice directions.
    """
    ax = abs(vx)
    ay = abs(vy)
    az = abs(vz)

    if ax >= ay and ax >= az:
        if vx >= 0.0:
            return 0
        return 1

    if ay >= ax and ay >= az:
        if vy >= 0.0:
            return 2
        return 3

    if vz >= 0.0:
        return 4
    return 5


@njit
def flat_index(x, y, z, N):
    return x * N * N + y * N + z


@njit
def speed_from_energy_relativistic_numba(W_erg):
    gamma = 1.0 + W_erg / M_E_C2
    beta2 = 1.0 - 1.0 / (gamma * gamma)

    if beta2 < 0.0:
        beta2 = 0.0
    if beta2 > 1.0 - 1e-14:
        beta2 = 1.0 - 1e-14

    return C * np.sqrt(beta2)


@njit
def find_next_event_numba(x0, y0, z0, dir_id, scatter_flat, N):
    dx_i, dy_i, dz_i = dir_int_components(dir_id)

    x = x0
    y = y0
    z = z0
    n_steps = 0

    while True:
        x += dx_i
        y += dy_i
        z += dz_i
        n_steps += 1

        if x < 0 or x >= N or y < 0 or y >= N or z < 0 or z >= N:
            return 0, x, y, z, n_steps

        if scatter_flat[flat_index(x, y, z, N)]:
            return 1, x, y, z, n_steps


@njit
def sample_cloud_dir_numba(vx, vy, vz, speed, use_rejection_int):
    if use_rejection_int == 0:
        return np.random.randint(0, 6)

    while True:
        cand = np.random.randint(0, 6)
        cx, cy, cz = dir_components(cand)

        Vx = V_A * cx
        Vy = V_A * cy
        Vz = V_A * cz

        rx = vx - Vx
        ry = vy - Vy
        rz = vz - Vz

        v_rel = np.sqrt(rx * rx + ry * ry + rz * rz)
        v_max = speed + V_A

        if np.random.random() <= v_rel / v_max:
            return cand


@njit
def scatter_random_lattice_outgoing_numba(W, speed, dir_id, use_rejection_int):
    """
    Generalized Eq.4 with random cloud-frame outgoing direction:

        dW ≈ (m c^2 + W1)/c^2 * [ V^2 - V · v1 + V · v2' ]

    Important:
        v2' is sampled in the cloud frame.
        The next lattice direction is NOT v2' directly.
        We transform back to lab frame:

            v2_lab = V + v2'

        and then project v2_lab to the nearest lattice direction.
    """
    dx, dy, dz = dir_components(dir_id)

    v1x = speed * dx
    v1y = speed * dy
    v1z = speed * dz

    cloud_id = sample_cloud_dir_numba(
        v1x, v1y, v1z, speed, use_rejection_int
    )

    cx, cy, cz = dir_components(cloud_id)

    Vx = V_A * cx
    Vy = V_A * cy
    Vz = V_A * cz

    u1x = v1x - Vx
    u1y = v1y - Vy
    u1z = v1z - Vz
    u1_mag = np.sqrt(u1x * u1x + u1y * u1y + u1z * u1z)

    # Cloud-frame outgoing direction
    out_id_prime = np.random.randint(0, 6)
    ox, oy, oz = dir_components(out_id_prime)

    v2px = u1_mag * ox
    v2py = u1_mag * oy
    v2pz = u1_mag * oz

    V2 = V_A * V_A
    V_dot_v1 = Vx * v1x + Vy * v1y + Vz * v1z
    V_dot_v2p = Vx * v2px + Vy * v2py + Vz * v2pz

    E_total = M_E_C2 + W

    dW_raw = E_total * (V2 - V_dot_v1 + V_dot_v2p) / (C * C)

    W_new = W + dW_raw
    if W_new < 0.0:
        W_new = 0.0

    actual_dW = W_new - W

    speed_new = speed_from_energy_relativistic_numba(W_new)
    if speed_new > 0.999999999 * C:
        speed_new = 0.999999999 * C

    dir_new = np.random.randint(0, 6)


    return W_new, speed_new, dir_new, actual_dW


# ============================================================
# 4. Numba simulation kernel
# ============================================================

@njit
def run_particles_numba(
    scatter_flat,
    time,
    pos_hist,
    E_hist,
    active_hist,
    E0,
    event_count,
    escape_time,
    escape_energy,
    event_frac_sample,
    event_dW_ev_sample,
    N_particles,
    N_grid,
    dx,
    T_final,
    sigma_v,
    use_rejection_int,
    seed,
):
    np.random.seed(seed)

    n_records = len(time)
    max_event_samples = len(event_frac_sample)
    event_sample_count = 0

    for p in range(N_particles):
        x = np.random.randint(0, N_grid)
        y = np.random.randint(0, N_grid)
        z = np.random.randint(0, N_grid)

        pos_hist[p, 0, 0] = x * dx
        pos_hist[p, 0, 1] = y * dx
        pos_hist[p, 0, 2] = z * dx
        active_hist[p, 0] = True

        vx0 = np.random.normal(0.0, sigma_v)
        vy0 = np.random.normal(0.0, sigma_v)
        vz0 = np.random.normal(0.0, sigma_v)

        speed = np.sqrt(vx0 * vx0 + vy0 * vy0 + vz0 * vz0)

        if speed < 1e-30:
            speed = 1e-30
        if speed > 0.999999 * C:
            speed = 0.999999 * C

        W = 0.5 * M_E * speed * speed
        E0[p] = W
        E_hist[p, 0] = W

        dir_id = np.random.randint(0, 6)

        t = 0.0
        rec_i = 1
        alive = True

        for ev in range(MAX_EVENTS_PER_PARTICLE):
            if not alive or t >= T_final:
                break

            hit, nx, ny, nz, n_steps = find_next_event_numba(
                x, y, z, dir_id, scatter_flat, N_grid
            )

            s = n_steps * dx
            dt = s / speed
            t_next = t + dt

            sx = x * dx
            sy = y * dx
            sz = z * dx

            ex = nx * dx
            ey = ny * dx
            ez = nz * dx

            t_fill_end = t_next
            if t_fill_end > T_final:
                t_fill_end = T_final

            while rec_i < n_records and time[rec_i] <= t_fill_end:
                tau = 0.0
                if dt > 0.0:
                    tau = (time[rec_i] - t) / dt

                if tau < 0.0:
                    tau = 0.0
                if tau > 1.0:
                    tau = 1.0

                rx = sx + tau * (ex - sx)
                ry = sy + tau * (ey - sy)
                rz = sz + tau * (ez - sz)

                inside = (
                    rx >= 0.0 and rx <= L_BOX
                    and ry >= 0.0 and ry <= L_BOX
                    and rz >= 0.0 and rz <= L_BOX
                )

                if inside:
                    pos_hist[p, rec_i, 0] = rx
                    pos_hist[p, rec_i, 1] = ry
                    pos_hist[p, rec_i, 2] = rz
                    E_hist[p, rec_i] = W
                    active_hist[p, rec_i] = True
                else:
                    active_hist[p, rec_i] = False

                rec_i += 1

            t = t_next
            x = nx
            y = ny
            z = nz

            if hit == 0:
                alive = False
                escape_time[p] = t
                escape_energy[p] = W
                break

            old_W = W

            W, speed, dir_id, actual_dW = scatter_random_lattice_outgoing_numba(
                W, speed, dir_id, use_rejection_int
            )

            event_count[p] += 1

            if event_sample_count < max_event_samples:
                if old_W > 0.0:
                    event_frac_sample[event_sample_count] = actual_dW / old_W
                else:
                    event_frac_sample[event_sample_count] = np.nan

                event_dW_ev_sample[event_sample_count] = actual_dW * ERG_TO_EV
                event_sample_count += 1

        while rec_i < n_records:
            if alive:
                inside_cell = (
                    x >= 0 and x < N_grid
                    and y >= 0 and y < N_grid
                    and z >= 0 and z < N_grid
                )

                if inside_cell:
                    pos_hist[p, rec_i, 0] = x * dx
                    pos_hist[p, rec_i, 1] = y * dx
                    pos_hist[p, rec_i, 2] = z * dx
                    E_hist[p, rec_i] = W
                    active_hist[p, rec_i] = True
                else:
                    active_hist[p, rec_i] = False
            else:
                active_hist[p, rec_i] = False

            rec_i += 1

    return event_sample_count


# ============================================================
# 5. Python wrapper for simulation
# ============================================================

def run_simulation_numba():
    rng = np.random.default_rng(SEED)

    dx = L_BOX / (N_GRID - 1)
    lambda_sc = dx / SCATTER_DENSITY

    dt_reference = 0.5 * dx / C
    rec_steps = np.arange(0, N_STEPS, RECORD_EVERY)
    if rec_steps[-1] != N_STEPS - 1:
        rec_steps = np.append(rec_steps, N_STEPS - 1)

    time = rec_steps * dt_reference
    T_FINAL = time[-1]
    n_records = len(time)

    print("=== Numba event-driven lattice simulation ===")
    print("SCATTER MODEL = generalized Eq.4; cloud-frame outgoing direction transformed back to lab")
    print(f"USE_REJECTION = {USE_REJECTION}")
    print(f"N_GRID        = {N_GRID}")
    print(f"dx            = {dx:.3e} cm")
    print(f"lambda_sc     = {lambda_sc:.3e} cm")
    print(f"T_FINAL       = {T_FINAL:.3f} s")
    print(f"N_RECORDS     = {n_records}")
    print(f"N_PARTICLES   = {N_PARTICLES}")
    print()

    print("Generating scatterer grid...")
    scatter_flat = make_scatter_grid_flat(N_GRID, SCATTER_DENSITY, rng)
    print(f"actual scatterer density = {scatter_flat.mean():.5f}")
    print()

    pos_hist = np.full((N_PARTICLES, n_records, 3), np.nan, dtype=np.float32)
    E_hist = np.full((N_PARTICLES, n_records), np.nan, dtype=np.float64)
    active_hist = np.zeros((N_PARTICLES, n_records), dtype=np.bool_)

    E0 = np.zeros(N_PARTICLES, dtype=np.float64)
    event_count = np.zeros(N_PARTICLES, dtype=np.int64)

    escape_time = np.full(N_PARTICLES, np.nan, dtype=np.float64)
    escape_energy = np.full(N_PARTICLES, np.nan, dtype=np.float64)

    event_frac_sample = np.full(MAX_EVENT_SAMPLES, np.nan, dtype=np.float64)
    event_dW_ev_sample = np.full(MAX_EVENT_SAMPLES, np.nan, dtype=np.float64)

    sigma_v = np.sqrt(T_INIT_EV * EV_TO_ERG / M_E)
    use_rejection_int = 1 if USE_REJECTION else 0

    print("Running particles with numba...")
    print("First run includes JIT compilation time.")

    event_sample_count = run_particles_numba(
        scatter_flat,
        time,
        pos_hist,
        E_hist,
        active_hist,
        E0,
        event_count,
        escape_time,
        escape_energy,
        event_frac_sample,
        event_dW_ev_sample,
        N_PARTICLES,
        N_GRID,
        dx,
        T_FINAL,
        sigma_v,
        use_rejection_int,
        SEED + 12345,
    )

    event_frac = event_frac_sample[:event_sample_count].copy()
    event_dW_ev = event_dW_ev_sample[:event_sample_count].copy()

    print("Done.")
    print(f"final active = {np.sum(active_hist[:, -1])}")
    print("event count percentiles:")
    print(np.percentile(event_count, [0, 10, 50, 90, 99, 99.9, 100]))

    if len(event_frac) > 0:
        finite_frac = event_frac[np.isfinite(event_frac)]
        finite_dW = event_dW_ev[np.isfinite(event_dW_ev)]

        print("\nSampled event fractional energy change actual dW/W percentiles:")
        print(np.percentile(finite_frac, [0, 1, 10, 50, 90, 99, 100]))
        print(f"sample mean actual dW/W = {np.mean(finite_frac):.3e}")

        print("\nSampled event absolute actual dW [eV] percentiles:")
        print(np.percentile(finite_dW, [0, 1, 10, 50, 90, 99, 100]))
        print(f"sample mean actual dW [eV] = {np.mean(finite_dW):.3e}")

    return (
        pos_hist,
        E_hist,
        E0,
        time,
        L_BOX,
        lambda_sc,
        active_hist,
        event_count,
        escape_time,
        escape_energy,
        event_frac,
        event_dW_ev,
    )


# ============================================================
# 6. Fig.3c escape-aware analysis
# ============================================================

def fig3c_escape_aware(
    pos_hist,
    E_hist,
    time,
    L_box,
    escape_energy,
    tmin,
    tmax,
    n_bins=16,
    min_bin=20,
):
    Ebin = np.where(
        np.isfinite(escape_energy),
        escape_energy,
        E_hist[:, -1],
    ) * ERG_TO_EV

    valid = np.isfinite(Ebin) & (Ebin > 0)

    print("\nEscape-aware Fig.3c:")
    print(f"valid particles = {np.sum(valid)} / {len(valid)}")

    if np.sum(valid) < min_bin:
        print("Too few particles.")
        return None

    E_valid = Ebin[valid]
    q = np.linspace(1.0, 99.5, n_bins + 1)
    bins = np.percentile(E_valid, q)
    bins = np.unique(bins)

    if len(bins) < 3:
        print("Too few unique energy bins.")
        return None

    n_bins_eff = len(bins) - 1
    centers = np.sqrt(bins[:-1] * bins[1:])

    alpha = np.full(n_bins_eff, np.nan)
    n_fixed = np.zeros(n_bins_eff, dtype=int)
    n_time_fit = np.zeros(n_bins_eff, dtype=int)
    r2_fit = np.full(n_bins_eff, np.nan)

    r0 = pos_hist[:, 0, :].astype(float)

    min_time_particles = max(30, min_bin)
    min_fit_points = 8

    for b in range(n_bins_eff):
        selected = valid & (Ebin >= bins[b]) & (Ebin < bins[b + 1])

        if b == n_bins_eff - 1:
            selected = valid & (Ebin >= bins[b]) & (Ebin <= bins[b + 1])

        ids = np.where(selected)[0]
        n_fixed[b] = len(ids)

        if len(ids) < min_bin:
            continue

        msd = np.full(len(time), np.nan)

        for i in range(len(time)):
            if time[i] < tmin or time[i] > tmax:
                continue

            dr = pos_hist[ids, i, :].astype(float) - r0[ids]
            ok = np.all(np.isfinite(dr), axis=1)

            if np.sum(ok) < min_time_particles:
                continue

            msd[i] = np.mean(np.sum(dr[ok] ** 2, axis=1))

        fit_mask = (
            (time >= tmin)
            & (time <= tmax)
            & (time > 0)
            & (msd > 0)
            & np.isfinite(msd)
        )

        n_time_fit[b] = np.sum(fit_mask)

        if n_time_fit[b] < min_fit_points:
            continue

        x = np.log(time[fit_mask])
        y = np.log(msd[fit_mask])

        slope, intercept = np.polyfit(x, y, 1)
        y_pred = intercept + slope * x

        ss_res = np.sum((y - y_pred) ** 2)
        ss_tot = np.sum((y - np.mean(y)) ** 2)

        if ss_tot > 0:
            r2 = 1.0 - ss_res / ss_tot
        else:
            r2 = np.nan

        if not np.isfinite(r2) or r2 < 0.95:
            continue

        alpha[b] = slope
        r2_fit[b] = r2

    for E, a, n, nt, r2 in zip(centers, alpha, n_fixed, n_time_fit, r2_fit):
        if np.isfinite(a):
            print(f"E={E:.3e} eV, a_r={a:.3f}, N={n}, Nt={nt}, R2={r2:.3f}")

    return {
        "centers": centers,
        "alpha": alpha,
        "n_fixed": n_fixed,
        "n_time_fit": n_time_fit,
        "r2_fit": r2_fit,
        "Ebin": Ebin,
        "bins": bins,
    }


# ============================================================
# 7. Global analysis and plotting
# ============================================================

def analyze(
    pos_hist,
    E_hist,
    E0,
    time,
    L_box,
    lambda_sc,
    event_count,
    escape_time,
    escape_energy,
    event_frac,
    event_dW_ev,
):
    Eev = E_hist * ERG_TO_EV
    E0ev = E0 * ERG_TO_EV

    in_box = (
        (pos_hist[:, :, 0] >= 0) & (pos_hist[:, :, 0] <= L_box)
        & (pos_hist[:, :, 1] >= 0) & (pos_hist[:, :, 1] <= L_box)
        & (pos_hist[:, :, 2] >= 0) & (pos_hist[:, :, 2] <= L_box)
    )

    r0 = pos_hist[:, 0, :].astype(float)

    spatial = np.full(len(time), np.nan)
    mean_dE = np.full(len(time), np.nan)
    msd_E = np.full(len(time), np.nan)
    active_number = np.zeros(len(time), dtype=int)

    for i in range(len(time)):
        mask = in_box[:, i] & np.isfinite(Eev[:, i])
        active_number[i] = np.sum(mask)

        if np.sum(mask) < 10:
            continue

        dr = pos_hist[mask, i, :].astype(float) - r0[mask]
        spatial[i] = np.mean(np.sum(dr**2, axis=1))

        dE = Eev[mask, i] - E0ev[mask]
        mean_dE[i] = np.mean(dE)
        msd_E[i] = np.mean(dE**2)

    a_r, fit_r = fit_power(time, spatial, TMIN_R, TMAX_FIT)
    a_W, fit_W = fit_power(time, mean_dE, TMIN_E, TMAX_FIT)
    a_W2, fit_W2 = fit_power(time, msd_E, TMIN_E, TMAX_FIT)

    print("\nGlobal fits:")
    print(f"a_r  = {a_r:.3f}")
    print(f"a_W  = {a_W:.3f}")
    print(f"a_W2 = {a_W2:.3f}")

    a_r_article, fit_r_article = fit_article_data(ARTICLE_SPATIAL_MSD)
    a_W_article, fit_W_article = fit_article_data(ARTICLE_MEAN_DW)
    a_W2_article, fit_W2_article = fit_article_data(ARTICLE_ENERGY_MSD)

    print("\nDigitized article fits:")
    print(f"article a_r  = {a_r_article:.3f}")
    print(f"article a_W  = {a_W_article:.3f}")
    print(f"article a_W2 = {a_W2_article:.3f}")

    res3c = fig3c_escape_aware(
        pos_hist=pos_hist,
        E_hist=E_hist,
        time=time,
        L_box=L_box,
        escape_energy=escape_energy,
        tmin=TMIN_R,
        tmax=TMAX_FIT,
        n_bins=N_BINS_3C,
        min_bin=MIN_BIN_3C,
    )

    Ebin = np.where(
        np.isfinite(escape_energy),
        escape_energy,
        E_hist[:, -1],
    ) * ERG_TO_EV

    print("\nFinal / escape energy percentiles [eV]:")
    print(np.nanpercentile(Ebin, [1, 10, 50, 90, 99, 99.9, 100]))

    print("\nActive particles at selected times:")
    for tt0 in [0, 1, 3, 5, 8, 10, time[-1]]:
        j = np.argmin(np.abs(time - tt0))
        print(f"t={time[j]:.2f} s: active={active_number[j]}")

    fig = plt.figure(figsize=(15, 10))
    mask_t = time >= 0.1

    ax1 = fig.add_subplot(2, 2, 1)
    ax1.loglog(time[mask_t], spatial[mask_t], "b.", ms=2, alpha=0.45, label="Simulation")

    tt = np.logspace(np.log10(TMIN_R), np.log10(TMAX_FIT), 200)
    ax1.loglog(tt, fit_r(tt), "r--", lw=2, label=fr"Simulation fit: $a_r={a_r:.2f}$")

    t_art = ARTICLE_SPATIAL_MSD[:, 0]
    y_art = ARTICLE_SPATIAL_MSD[:, 1]
    ax1.loglog(t_art, y_art, "ko", ms=4, alpha=0.85, label="Article data")

    tt_art = np.logspace(np.log10(t_art.min()), np.log10(t_art.max()), 200)
    ax1.loglog(tt_art, fit_r_article(tt_art), "k--", lw=2, label=fr"Article fit: $a_r={a_r_article:.2f}$")

    alpha_article_label = 1.41
    log_A_label = np.mean(np.log(y_art) - alpha_article_label * np.log(t_art))
    A_article_label = np.exp(log_A_label)

    ax1.loglog(
        tt_art,
        A_article_label * tt_art**alpha_article_label,
        color="gray",
        linestyle="-.",
        lw=2,
        label=fr"article label line: $\alpha_r=1.41$",
    )

    ax1.set_title("Spatial MSD")
    ax1.set_xlabel("Time [s]")
    ax1.set_ylabel(r"$\langle(\Delta r)^2\rangle$ [cm$^2$]")
    ax1.legend(fontsize=9)

    ax2 = fig.add_subplot(2, 2, 2)
    ax2.loglog(time[mask_t], msd_E[mask_t], "g.", ms=2, alpha=0.45, label="Simulation")

    ttE = np.logspace(np.log10(TMIN_E), np.log10(TMAX_FIT), 200)
    ax2.loglog(ttE, fit_W2(ttE), "r--", lw=2, label=fr"Simulation fit: $a_{{W2}}={a_W2:.2f}$")

    t_art = ARTICLE_ENERGY_MSD[:, 0]
    y_art = ARTICLE_ENERGY_MSD[:, 1]
    ax2.loglog(t_art, y_art, "ko", ms=4, alpha=0.85, label="Article data")

    tt_art = np.logspace(np.log10(t_art.min()), np.log10(t_art.max()), 200)
    ax2.loglog(tt_art, fit_W2_article(tt_art), "k--", lw=2, label=fr"Article fit: $a_{{W2}}={a_W2_article:.2f}$")

    ax2.set_title("Energy MSD")
    ax2.set_xlabel("Time [s]")
    ax2.set_ylabel(r"$\langle(\Delta W)^2\rangle$ [eV$^2$]")
    ax2.legend(fontsize=9)

    ax3 = fig.add_subplot(2, 2, 3)
    pos_mask = mask_t & (mean_dE > 0)
    ax3.loglog(time[pos_mask], mean_dE[pos_mask], "m.", ms=2, alpha=0.45, label="Simulation")

    ax3.loglog(ttE, fit_W(ttE), "r--", lw=2, label=fr"Simulation fit: $a_W={a_W:.2f}$")

    t_art = ARTICLE_MEAN_DW[:, 0]
    y_art = ARTICLE_MEAN_DW[:, 1]
    ax3.loglog(t_art, y_art, "ko", ms=4, alpha=0.85, label="Article data")

    tt_art = np.logspace(np.log10(t_art.min()), np.log10(t_art.max()), 200)
    ax3.loglog(tt_art, fit_W_article(tt_art), "k--", lw=2, label=fr"Article fit: $a_W={a_W_article:.2f}$")

    ax3.set_title("Mean energy gain")
    ax3.set_xlabel("Time [s]")
    ax3.set_ylabel(r"$\langle\Delta W\rangle$ [eV]")
    ax3.legend(fontsize=9)

    ax4 = fig.add_subplot(2, 2, 4)

    if res3c is not None:
        E = res3c["centers"]
        alpha = res3c["alpha"]
        ntime = res3c["n_time_fit"]

        good = np.isfinite(alpha) & (ntime > 0)

        if np.any(good):
            sc = ax4.scatter(
                E[good],
                alpha[good],
                c=ntime[good],
                s=45,
                alpha=0.9,
            )
            cb = fig.colorbar(sc, ax=ax4)
            cb.set_label("Number of fitted time points")

    ax4.axhline(1.0, color="k", ls="--", lw=1.2, label=r"$a_r=1$")
    ax4.axhline(1.6, color="gray", ls=":", lw=1.5, label=r"$a_r=1.6$")
    ax4.axhline(2.0, color="red", ls="-.", lw=1.2, label=r"$a_r=2$")

    ax4.set_xscale("log")
    ax4.set_title("Fig.3c-like escape-aware")
    ax4.set_xlabel("Final / escape energy [eV]")
    ax4.set_ylabel(r"$a_r$")
    ax4.legend(fontsize=9)

    plt.tight_layout()
    plt.show()

    fig2 = plt.figure(figsize=(13, 4))

    ax = fig2.add_subplot(1, 3, 1)
    ax.plot(time, active_number, "k-")
    ax.set_xlabel("Time [s]")
    ax.set_ylabel("Active particles")
    ax.set_title("Surviving particles")

    ax = fig2.add_subplot(1, 3, 2)
    ax.hist(event_count, bins=60, histtype="step", color="k")
    ax.set_xlabel("Number of scatterings")
    ax.set_ylabel("Count")
    ax.set_title("Event count")

    finite_dW = event_dW_ev[np.isfinite(event_dW_ev)]
    ax = fig2.add_subplot(1, 3, 3)
    ax.hist(finite_dW, bins=100, histtype="step", color="k")
    ax.set_xlabel(r"$\Delta W$ [eV]")
    ax.set_ylabel("Count")
    ax.set_title("Sampled event actual energy change")
    ax.set_yscale("log")

    plt.tight_layout()
    plt.show()

    return {
        "a_r": a_r,
        "a_W": a_W,
        "a_W2": a_W2,
        "article_a_r": a_r_article,
        "article_a_W": a_W_article,
        "article_a_W2": a_W2_article,
        "fig3c": res3c,
        "spatial_msd": spatial,
        "mean_dE": mean_dE,
        "energy_msd": msd_E,
        "active_number": active_number,
        "event_count": event_count,
        "escape_time": escape_time,
        "escape_energy": escape_energy,
        "event_frac": event_frac,
        "event_dW_ev": event_dW_ev,
    }


# ============================================================
# 8. Main
# ============================================================

if __name__ == "__main__":
    (
        pos_hist,
        E_hist,
        E0,
        time,
        L_box,
        lambda_sc,
        active_hist,
        event_count,
        escape_time,
        escape_energy,
        event_frac,
        event_dW_ev,
    ) = run_simulation_numba()

    results = analyze(
        pos_hist=pos_hist,
        E_hist=E_hist,
        E0=E0,
        time=time,
        L_box=L_box,
        lambda_sc=lambda_sc,
        event_count=event_count,
        escape_time=escape_time,
        escape_energy=escape_energy,
        event_frac=event_frac,
        event_dW_ev=event_dW_ev,
    )
```
